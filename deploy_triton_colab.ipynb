```python
# Run in Google Colab with GPU runtime
import numpy as np
import pickle
from PIL import Image
import torchvision.transforms as transforms
import time
import tritonclient.http as triton_http
import os
import onnxruntime as ort



# Simulated 100 logs (same as CPU)
new_logs = ["Script crashed" + str(i) if i % 3 == 0 else "Typo in code" + str(i) if i % 3 == 1 else "High CPU" + str(i) for i in range(100)]
new_severity = [2.9 if i % 3 == 0 else 0.5 if i % 3 == 1 else 1.8 for i in range(100)]
new_image_paths = ['error.png' if i % 3 == 0 else 'typo.png' if i % 3 == 1 else 'warning.png' for i in range(100)]

# Preprocess text
vectorizer = pickle.load(open('tfidf_vectorizer.pkl', 'rb'))
text_features = vectorizer.transform(new_logs).toarray()  # (100, 46)

# Preprocess images
transform = transforms.ToTensor()
image_features = []
for path in new_image_paths:
    img = Image.open(path)
    img_tensor = transform(img)
    img_flat = img_tensor.view(-1).numpy()
    image_features.append(img_flat)
image_features = np.array(image_features)  # (100, 2352)

# Combine features
all_features = np.hstack((text_features, image_features))  # (100, 2398)

# Load JAX important features
top_indices = np.load('important_features.npy')  # e.g., [1875, 561, ...]
jax_features = all_features[:, top_indices]  # (100, 20), not used yet

# Setup Triton server (simplified, assumes severity_model.onnx uploaded to Colab)
# Note: Manual setup required, see instructions below
# 1. Upload severity_model.onnx to Colab
# 2. Create model repository: /models/severity_model/1/model.onnx
# 3. Run Triton server: !tritonserver --model-repository=/models
# For simplicity, using ONNX Runtime as fallback if Triton setup fails
try:
    client = triton_http.InferenceClient(url='localhost:8000')  # Adjust if needed
    inputs = triton_http.InferInput('dense_input', all_features.shape, 'FP32')
    inputs.set_data_from_numpy(all_features.astype(np.float32))
    start_time = time.time()
    result = client.infer(model_name='severity_model', inputs=[inputs])
    predictions_all = result.as_numpy('output')  # Adjust 'output' to model
    time_all = time.time() - start_time
except:
    print("Triton server setup failed, using ONNX Runtime fallback")
    import onnxruntime as ort
    session = ort.InferenceSession('severity_model.onnx')
    input_name = session.get_inputs()[0].name
    start_time = time.time()
    predictions_all = session.run(None, {input_name: all_features.astype(np.float32)})[0]
    time_all = time.time() - start_time

# Compute metrics
mse_all = np.mean((predictions_all.flatten() - new_severity) ** 2)
mae_all = np.mean(np.abs(predictions_all.flatten() - new_severity))

# Note: JAX features require retrained ONNX model
print("All Features (2398):")
print(f"Test MSE: {mse_all:.4f}, Test MAE: {mae_all:.4f}, Time: {time_all:.4f}s")
print("JAX Features (20): Not tested (requires retrained model with 20 inputs)")
print("Sample predictions (first 5, all features):")
for i in range(5):
    print(f"Log {i+1}: {new_logs[i]} | True: {new_severity[i]:.1f} | Predicted: {predictions_all[i][0]:.3f}")
```
